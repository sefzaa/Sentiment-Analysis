```python
# Mulai crawling search tweet
tweets = tw.Cursor(api.search_tweets, q="kapolri OR #kapolri OR #listyosigitprabowo OR #listyosigit -filter:retweets",
                      lang="id", tweet_mode='extended').items(500)
# Membuat list 
data_search =[]

# melakukakn 
for teks in tweets:
    #Membuat dict 
    teks_properties= {}
    #Membuat properti tanggal, nama pengguna dan isi tweet yang telah dibersihkan
    teks_properties["Tanggal"]=teks.created_at
    teks_properties["Pengguna"]=teks.user.screen_name
    teks_clean = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",teks.full_text).split())
    teks_properties["Tweet"]= teks_clean
    
    #Memamsukkan fungsi textBlob dengan argumen isi tweet ke variabel analysis
    analysis = TextBlob(teks_clean)
    #Metranslate isi tweet bahasa indonesia ke bahasa inggris
    analysis = analysis.translate(from_lang='id', to='en')
    #Menentukan apakah tweet merupakan sentimen positif, negatif dan netral berdasarkan polaritasnya
    if analysis.sentiment.polarity > 0.0:
      teks_properties["Sentimen"]= 'Positif'
    elif analysis.sentiment.polarity == 0.0:
      teks_properties["Sentimen"]= 'Netral'
    else:
      teks_properties["Sentimen"]= 'Negatif'

    #Menghilangkan retweet
    if teks.retweet_count > 0:
        if teks_properties not in data_search:
            data_search.append(teks_properties)
    else:
        data_search.append(teks_properties)
```

```python
# Membagi/split tweet menjadi positif negatif dan netral
tweet_positif = [t for t in data_search if t['Sentimen'] == 'Positif']
tweet_netral = [t for t in data_search if t['Sentimen'] == 'Netral']
tweet_negatif = [t for t in data_search if t['Sentimen'] == 'Negatif']
```

```python
# Menentukan jumlah tweet tiap sentimen
print('hasil sentimen')
print('positif ', len(tweet_positif), '({:.2f}%)'.format(100*len(tweet_positif)/len(data_search)))
print('netral ', len(tweet_netral), '({:.2f}%)'.format(100*len(tweet_netral)/len(data_search)))
print('negatif ', len(tweet_negatif), '({:.2f}%)'.format(100*len(tweet_negatif)/len(data_search)))
```
